```python
# Temel Sınıf: Animal
class Animal:
    def __init__(self, species):
        self.species = species

    def show_info(self):
        print(f"Species: {self.species}")

# Mammal Sınıfı
class Mammal(Animal):
    def __init__(self, species, weight, speed):
        super().__init__(species)
        self.weight = weight
        self.speed = speed

    def show_info(self):
        super().show_info()
        print(f"Weight: {self.weight} kg, Speed: {self.speed} km/h")

# Bird Sınıfı
class Bird(Animal):
    def __init__(self, species, wingspan):
        super().__init__(species)
        self.wingspan = wingspan

    def show_info(self):
        super().show_info()
        print(f"Wingspan: {self.wingspan} m")

# Reptile Sınıfı
class Reptile(Animal):
    def __init__(self, species, length):
        super().__init__(species)
        self.length = length

    def show_info(self):
        super().show_info()
        print(f"Length: {self.length} m")

# Zookeeper Sınıfı
class Zookeeper:
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id
        self.animals = []

    def assign_animal(self, animal):
        self.animals.append(animal)

    def show_animals(self):
        print(f"Zookeeper: {self.name}, ID: {self.employee_id}")
        for animal in self.animals:
            animal.show_info()
            print()

# Zoo Sınıfı
class Zoo:
    def __init__(self):
        self.zookeepers = []

    def add_zookeeper(self, zookeeper):
        self.zookeepers.append(zookeeper)

    def show_zookeepers(self):
        for zookeeper in self.zookeepers:
            zookeeper.show_animals()

# Kullanım
lion = Mammal("Lion", 190, 80)
eagle = Bird("Eagle", 2.1)
snake = Reptile("Snake", 3)

john = Zookeeper("John", "ZK001")
john.assign_animal(lion)
john.assign_animal(eagle)

zoo = Zoo()
zoo.add_zookeeper(john)

zoo.show_zookeepers()
```
